FEATURE ENGINEERING

```python
df['HR_diff'] = df['HR'].diff()
# df[vitals] = df[vitals].diff()
df['HR_diff'] = df.groupby('Patient_ID')['HR'].diff().fillna(0)
df['Resp_diff'] = df.groupby('Patient_ID')['HR'].diff().fillna(0)
df['MAP_diff'] = df.groupby('Patient_ID')['HR'].diff().fillna(0)
```

the above code will show how much HR value changed from previous hour.

```python
df['HR_roll_mean'] = df['HR'].rolling(3).mean().fillna(0)
df['MAP_roll_std'] = df['MAP'].rolling(3).std().fillna(0)
```

```python
df.shape
```

```python
null_counts = df.isna().sum()
print(null_counts[null_counts > 0])
```

```python
df['HR_Temp'] = df['HR'] * df['Temp']
```

Combine two signals.
Ex:
fever + tachycardia together stronger than individually

```python
df['Pulse_Pressure'] = df['SBP'] - df['DBP']
```

PULSE PRESSURE: Low pulse pressure may indicate:

poor circulation
shock
cardiac dysfunction

```python
df['Shock_Index'] = df['HR'] / df['SBP']
```

SHOCK INDEX : In shock/sepsis:

HR rises
BP falls

So shock index increases.

```python
threshold = 0.95

missing_ratio = df.isnull().mean()

drop_cols = missing_ratio[
    missing_ratio > threshold
].index

df = df.drop(columns=drop_cols)
```

```python
df['HospAdmTime'] = df['HospAdmTime'].fillna(
    df['HospAdmTime'].median()
)
```

```python
df['Unit1'] = df['Unit1'].fillna(0)
df['Unit2'] = df['Unit2'].fillna(0)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['DBP'] = (
    df.groupby('Patient_ID')['DBP']
      .transform(lambda x: x.interpolate(limit=2))
)

df['DBP'] = (
    df.groupby('Patient_ID')['DBP']
      .ffill()
)

df['DBP'] = df['DBP'].fillna(df['DBP'].median())
```

```python
df = df.drop('BaseExcess',axis=1)
```

```python
drop_missing_cols = [
    'Unnamed: 0_missing',
    'Hour_missing',
    'Age_missing',
    'Gender_missing',
    'ICULOS_missing',
    'SepsisLabel_missing',
    'Patient_ID_missing',
    'HospAdmTime_missing'
]
```

```python
df=df.drop(drop_missing_cols,axis=1)
```

```python
df = df.drop(columns=['Unit1_missing','Unit2_missing','Unnamed: 0'],axis=1)
```

```python
important_missing = ['FiO2', 'pH', 'PaCO2']

for col in important_missing:
    df[col + '_missing'] = df[col].isnull().astype(int)


important_labs = ['FiO2', 'pH', 'PaCO2', 'Hct']

df[important_labs] = (
    df.groupby('Patient_ID')[important_labs]
      .ffill()
)


for col in important_labs:
    df[col] = df[col].fillna(df[col].median())

remaining_nulls = df.isnull().sum()

print(
    remaining_nulls[remaining_nulls > 0]
        .sort_values(ascending=False)
)


print(df.shape)
```
